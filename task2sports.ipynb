```python
import pandas as pd
import numpy as np
from numpy import mean
from scipy import stats
from statistics import mode as md
df = pd.read_csv('./dataset_Facebook.csv', sep=';')
df
```

```python
def get_mean_all(array_value):
    for each_column in array_value:
        if (each_column != 'Type'): 
            print('Mean of ' + str(each_column) + ' = ' + str(df[each_column].mean()))

def get_median_all(array_value):
    for each_column in array_value:
        if (each_column != 'Type'): 
            print('Median of ' + str(each_column) + ' = ' + str(df[each_column].median()))

def get_mode_all(array_value):
    for each_column in array_value:
        if (each_column != 'Type'): 
            print('Mode of ' + str(each_column) + ' = ' + str(stats.mode(df[each_column]).mode))

def get_max_all(array_value):
    for each_column in array_value:
        if (each_column != 'Type'): 
            print('Max of ' + str(each_column) + ' = ' + str(max(df[each_column])))
            
def get_min_all(array_value):
    for each_column in array_value:
        if (each_column != 'Type'): 
            print('Min of ' + str(each_column) + ' = ' + str(min(df[each_column])))
```

```python
names_column = df.columns.tolist()

get_mean_all(names_column)
get_median_all(names_column)
get_mode_all(names_column)
get_max_all(names_column)
get_min_all(names_column)
```

```python
#Найдем уникальные значения типов
list_type_set = set(df['Type']) 
unique_type_list = (list(list_type_set)) 
print(list_type_set)
```

```python
#Профильтруем для каждого типа
names_column_array = []

for value_type in list_type_set:
    filtered = df.loc[df['Type'] == value_type]
    names_column_array.append(filtered)

for list_type in list_type_set:
    print('\nType of ' + list_type + '\n')
    
    for name_array in names_column_array:
        get_mean_all(name_array)
        get_median_all(name_array)
        get_mode_all(name_array)
        get_max_all(name_array)
        get_min_all(name_array)
```

```python
def normalize(x):
    return (x-x.min())/(x.max()-x.min())

new_df = df
new_df['par_1'] = (df['Lifetime Engaged Users'] / df['Lifetime Post Total Reach']).fillna(0)
new_df['par_2'] = (df['Lifetime Post Consumers'] / df['Lifetime Post Consumptions']).fillna(0)
new_df['par_3'] = (df['Lifetime Engaged Users'] / df['Lifetime Post Total Impressions']).fillna(0)
new_df['par_4'] = (0.35 * df['comment'] / df['Total Interactions']).fillna(0)
new_df['par_5'] = (0.75 * df['like'] / df['Total Interactions']).fillna(0)
new_df['par_6'] = (df['share'] / df['Total Interactions']).fillna(0)
```
